## Answering questions about the data

```python
# What's the time range convered in the dataset ?
min_date = olpics_df["Year"].min()
max_date = olpics_df["Year"].max()

f"The time ranges from {min_date} et {max_date}"
```

```python
# The Olympics take place every fours years. Why are there missing years?
"Because WWI from 1914 - 1918 and WWII from 1939 - 1945"
```

```python
# What are the type of medals awarded ?
olpics_df["Medal"].unique().tolist()
```

```python
# Across all of the Olympic Games, how many gold, silver, and bronze medals have been there?
olpics_df["Medal"].value_counts()
```

```python
# Why are there not an equel number of gold, silver, and bronze medals ?
olpics_df["Position"].value_counts()
```

```python
# There are more gold medals than silver and more silver than bronze. Why might that be ?
olpics_df["Position"].value_counts()
```

```python
# what are the different Notional Olympic Committees (NOCs) ?
olpics_df["NOC"].unique()
```

```python
# What does the NOC "ZZX" represent ?
"The mixed team"
```
