## Memory usage of a DataTypes

- The `category` data type is more memory efficient for columns with a limited number of unique values compared to using the `object` data type.
Mathematically, the memory savings can be significant, especially for large datasets with many repeated values.
- Furtthermore, using the `category` data type can lead to performance improvements in certain operations, such as filtering and grouping, because pandas can leverage the underlying integer representation of the categories.
- Also the `string` data type is optimized for text data and can provide better performance and memory efficiency compared to the `object` data type, which is a more general-purpose type that can hold any Python object.


```python
# changing all the non categorical object columns to 'string' data type
object_cols = olpics_df.select_dtypes(include="object").columns.tolist()
for col in object_cols:
    olpics_df[col] = olpics_df[col].astype("string")
    
olpics_df.dtypes
```

## Main difference between .loc and .iloc

- `.loc` is *label-based*, which means that you have to specify the names of the rows and columns that you want to select.
- `.iloc` is integer *position-based*, which means that you have to specify the integer indexes of the rows and columns that you want to select.

```python
# First case: Concatenating DataFrames vertically (stacking rows)
# Imaginez deux relevés de ventes mensuels
ventes_janvier = pd.DataFrame({
    'Produit': ['Pomme', 'Orange'],
    'Quantite': [10, 5]
})

ventes_fevrier = pd.DataFrame({
    'Produit': ['Banane', 'Poire'],
    'Quantite': [7, 3]
})
print(f"Ventes Janvier: {ventes_janvier}\n")
print(f"Ventes Février: {ventes_fevrier}\n")
```

```python
# On veut juste tout mettre dans un seul tableau
# axis=0 signifie "empiler verticalement" (ajouter des lignes)
total_ventes = pd.concat([ventes_janvier, ventes_fevrier], axis=0)
total_ventes
```

```python
pd.merge?
```

```python
# Second case: Merging DataFrames based on a common column
# Notez : Client 3 (Charlie) n'a PAS de commande.
df_clients = pd.DataFrame({
    'client_id': [1, 2, 3],
    'nom': ['Alice', 'Bob', 'Charlie']
})

# Table des Commandes
# Notez : Commande 105 a un client_id (99) qui n'existe PAS dans la table clients.
df_commandes = pd.DataFrame({
    'commande_id': [101, 102, 103, 104, 105],
    'client_id': [1, 1, 2, 1, 99], 
    'montant': [250, 100, 50, 300, 999]
})

print(f"Clients:\n{df_clients}\n")
print(f"Commandes:\n{df_commandes}\n")
```

```python
# --- A. INNER MERGE (L'Intersection) ---
# Garde SEULEMENT les clés présentes dans les DEUX tables.
# Charlie (3) disparaît car il n'a pas commandé.
# La commande 105 (99) disparaît car le client n'existe pas.
print("\n--- A. INNER JOIN (Intersection) ---")
merge_inner = pd.merge(df_commandes, df_clients, on='client_id', how='inner')
print(merge_inner)

# --- B. LEFT MERGE (Priorité à Gauche : Commandes) ---
# Garde TOUTES les commandes.
# Si le client n'est pas trouvé (commande 105), le nom sera NaN (vide).
print("\n--- B. LEFT JOIN (Tout de 'Commandes') ---")
merge_left = pd.merge(df_commandes, df_clients, on='client_id', how='left')
print(merge_left)

# --- C. RIGHT MERGE (Priorité à Droite : Clients) ---
# Garde TOUS les clients.
# Charlie (3) apparaît, même sans commande (montant sera NaN).
# La commande 105 disparaît car son client n'est pas dans la liste de droite.
print("\n--- C. RIGHT JOIN (Tout de 'Clients') ---")
merge_right = pd.merge(df_commandes, df_clients, on='client_id', how='right')
print(merge_right)

# --- D. OUTER MERGE (L'Union Complète) ---
# Garde TOUT le monde.
# Charlie est là. La commande orpheline 105 est là aussi.
# Personne n'est laissé pour compte.
print("\n--- D. OUTER JOIN (Union Totale) ---")
merge_outer = pd.merge(df_commandes, df_clients, on='client_id', how='outer')
print(merge_outer)
```
