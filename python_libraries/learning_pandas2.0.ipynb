## Working with data types

```python
olpics_df.dtypes
```

```python
# Changeing all categorial columns to 'category' data type
categorical_cols = ["Gender", "Event_Gender", "Medal"]
for col in categorical_cols:
    olpics_df[col] = olpics_df[col].astype("category")

olpics_df.dtypes
```

```python
# changing the order of the 'Medal' category. That's is Bronze < Silver < Gold. Because by default it's in alphabetical order and in the Olympics, Bronze is less than Silver which is less than Gold.
medal_categories = ["Bronze", "Silver", "Gold"]
olpics_df["Medal"] = olpics_df["Medal"].cat.reorder_categories(medal_categories, ordered=True)
# Alternatively, we can also use the 'Categorical' data type constructor to achieve the same
# olpics_df["Medal"] = pd.Categorical(olpics_df["Medal"], categories=medal_categories, ordered=True)
olpics_df.dtypes
```
