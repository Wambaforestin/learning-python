## Stratégie 2 : Remplir les valeurs manquantes (souvent mieux)
> .fillna() est mon meilleur ami pour cela

````python
# Remplir le 'Product' manquant par "Inconnu"
df["Product"] = df["Product"].fillna("Inconnu")
````

## Remplir les valeurs manquantes dans 'Quantity Ordered' et 'Price Each' par la moyenne de leurs colonnes respectives

- 'pd.to_numeric' est l'outil parfait.
- 'errors='coerce'' est la clé : il transforme tout ce qu'il ne peut pas convertir (ex: "erreur", "N/A") en 'NaN' (Not a Number).

````python
# conversion des colonnes en numérique (analyse invalide -> NaN), calcul des moyennes, puis rempl
quantite_commande_num = pd.to_numeric(df["Quantity Ordered"], errors="coerce")
prix_num = pd.to_numeric(df["Price Each"], errors="coerce")

quantite_commande_moyenne = quantite_commande_num.mean()
prix_moyen = prix_num.mean()

# Remplir les valeurs manquantes dans 'Quantity Ordered' et 'Price Each' par la moyenne de leurs colonnes respectives
df["Quantity Ordered"] = quantite_commande_num.fillna(quantite_commande_moyenne).round().astype(int) # arrondi et conversion en int
df["Price Each"] = prix_num.fillna(prix_moyen)
````

## Valeurs manquentes après le nettoyage

````python
print("\nValeurs manquentes après le nettoyage:")
print(df.isnull().sum())
````

## NOTE IMPORTANTE SUR info() vs describe()

> En résumé : info() regarde la STRUCTURE, tandis que describe() regarde le CONTENU.

*Voici la différence détaillée :*

**1. df.info() : La Carte d'Identité**

C'est une analyse technique. Elle ne regarde pas vraiment les valeurs à l'intérieur, mais plutôt comment le tableau est construit en mémoire.

Ce qu'elle affiche :

- Le nombre de lignes et de colonnes.
    
- Le nom de chaque colonne.

- Le type de données (Dtype) de chaque colonne (est-ce du texte object, des entiers int64, des décimaux float64 ?).

- Le nombre de valeurs non-nulles (très utile pour repérer les trous dans les données).

- La quantité de mémoire vive (RAM) utilisée par le DataFrame.

*Quand l'utiliser : Dès le chargement du fichier, pour vérifier que les dates ne sont pas en texte, ou pour voir s'il manque beaucoup de données.*

**2. df.describe() : Le Bilan de Santé**

C'est une analyse statistique. Elle plonge dans les valeurs numériques pour vous donner une idée de leur distribution.

Ce qu'elle affiche (par défaut sur les colonnes numériques) :

- count : Nombre d'enregistrements.

- mean : La moyenne.

- std : L'écart-type (la dispersion des données). 
- min / max : Les valeurs minimales et maximales.

- 25% / 50% / 75% : Les quartiles (le 50% est la médiane).

*Quand l'utiliser : Pour comprendre l'ordre de grandeur de vos données. (Exemple : "Quel est le salaire moyen ?", "Est-ce qu'il y a des prix négatifs aberrants ?", "Est-ce que les âges sont bien compris entre 18 et 99 ?").*
